```python
%%file word_freq.py
from mrjob.job import MRJob
import re

WORD_RE = re.compile(r"^[bB][A-Za-z0-9_]*")



class MRWordFreqCount(MRJob):
    def testing(self,string1):
      cases=WORD_RE.findall(string1)
      return cases
    
    def mapper(self, _, line):
        # Your code here
        cases1=[]
        for x in line.split():
          cases1+=self.testing(x)
        yield "counts", len(cases1)

    def reducer(self, key, value):
        # Your code here
        #values1+=value
        yield key, sum(value)


if __name__ == '__main__':
    MRWordFreqCount.run()
```

```python
import word_freq

mr_job = word_freq.MRWordFreqCount(args=['alice.txt'])
with mr_job.make_runner() as runner:
     runner.run()
     for key, value in mr_job.parse_output(runner.cat_output()):
       print(key, value)
```
